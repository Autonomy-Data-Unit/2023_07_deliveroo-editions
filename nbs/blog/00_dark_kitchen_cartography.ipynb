# Dark Kitchen Cartography

![](assets/cover.png)

Dedicated sites for preparing takeaway meals for delivery only, known as "dark kitchens", have become a widely adopted model in the food service industry throughout the past half decade. The separation of a restaurant's delivery business from it's in-house business enables the delivery business to be operated from non-residential central locations where overhead and operating costs are reduced. Researching the successes and failures of this service model could provide useful insights for economic planning, risk management and food distribution, however there is a lack of detailed data on the growth and current state of dark kitchens within the UK. 

To the best of our knowledge, a map of the UK's dark kitchen locations does not exist. Therefore we set out to make one for Deliveroo's UK dark kitchen facilities, known as 'Editions', to better understand their growth and current state.

## Locating Deliveroo's Editions 



Deliveroo Editions was launched in May 2017 as a 'concept that puts an end to postcode food envy'. The concept was trialed in Camberwell, Battersea, Dulwich and Canary Wharf before scaling to [30 locations across the UK](https://www.wired.co.uk/article/deliveroo-editions-local-restaurants) within the first year. The Internet Archive's Wayback Machine maintains a capture of the webpage for the original Deliveroo Editions announcement.

![](assets/editions_launch.png)

Deliveroo disloses to users which restaurants are based at Editions facilities within the platform. We can filter restaurant search results for Editions partners by adding `tags=deliveroo+editions` as a query string in our URL when searching for restaurants within a given location. For example to find all the Editions partners delivering to Stepney Green, we can try `https://deliveroo.co.uk/restaurants/london/stepney-green?fulfillment_method=DELIVERY&tags=deliveroo+editions`:

![](assets/query.png)

If we check out Popeyes Louisiana Kitchen which is an Editions partner, we can find the name Deliveroo has given the Editions site from the restaurant's url: `https://deliveroo.co.uk/menu/London/whitechapel-editions/popeyes-editions-wme/?day=today&geohash=gcpvngtckn3c&time=ASAP` which is "Whitechapel Editions" in this case. We can conclude that Popeyes Louisiana Kitchen is also the address for Whitechapel Editions which happens to be 10 Assembly Passage, London, E14UT, a [2-storey B-class office space with a disco ball included built in 1906](https://www.loopnet.co.uk/Listing/10-40-Assembly-Passage-London/8354101/) right next to the busy A11 road into central London.

![](assets/assembly.jpeg)

To find all the other restaurants based at the Whitechapel Editions kitchen, we could search for editions restaurants near the editions name: `https://deliveroo.co.uk/restaurants/london/whitechapel-editions?fulfillment_method=DELIVERY&category=deliveroo+editions`. We can deduce that all the restaurants based 0.0 miles away are those at the editions kitchen and can validate this by checking for `whitechapel-editions` in their urls.
